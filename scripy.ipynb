```python
import networkx as nx
import matplotlib.pyplot as plt

def create_graph_from_adjacency_matrix(adj_matrix):
    G = nx.Graph()
    
    num_nodes = len(adj_matrix)
    G.add_nodes_from(range(num_nodes))
    
    for i in range(num_nodes):
        for j in range(i + 1, num_nodes):  # Check only upper triangle for undirected graph
            if adj_matrix[i][j] == 1:
                G.add_edge(i, j)
    
    return G
```

```python
# Example adjacency matrix
adj_matrix = [[0,1,1,1],[1,0,0,0],[1,0,0,0],[1,0,0,0]]

# Create the graph
graph = create_graph_from_adjacency_matrix(adj_matrix)

# Visualize the graph
plt.figure(figsize=(4, 4))
nx.draw(graph, with_labels=True, node_color='skyblue', node_size=1000, font_weight='bold', edge_color='gray')
plt.title("Graph ")
plt.show()
```
